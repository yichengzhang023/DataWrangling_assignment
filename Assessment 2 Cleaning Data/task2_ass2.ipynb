## 3. Resolving schema conflicts and merging data

### 3.1. Resolving schema conflicts

**From the above checking, we can find some schema conflicts, we should fix the conflicts column by column**

**First, fix data type/namespace conflicts**

```python
print(dataset1.ContractType.value_counts())
print(dataset2['Contract Type'].value_counts())
```

```python
print(dataset1.ContractTime.value_counts())
print(dataset2['Contract Time'].value_counts())
```

```python
print(dataset1.Category.value_counts())
print(dataset2.Category.value_counts())
```

There are conflicts in some of the columns:
* **Contract Type** 

null values replace by `non_specified`, format(ft -> full_time,pt -> part_time).
- - -
* **Contract Time** 

null values replace by `non_specified`, format(perm. -> permanent,contr. -> contract)
- - -
* **Category**

'Accounting & Finance Jobs' -> 'Finance & Accounting Jobs'
- - -
* **Salary per month**

It should be cauclated as salary per anniversary. Guess the relationship between salary per month and salary per anniversary should be 12 times(12 month).
- - -
* **Date**

The date format in dataset2 is `2012-04-06 12:00:00`, we should transfer this format to dataset1's format:`20120406T120000`

**Change the contract columns**

```python
replacetype = {'ft' : 'full_time','pt' : 'part_time'}
dataset2['Contract Type'].replace(replacetype,inplace = True)
dataset2['Contract Type'].fillna('non_specified',inplace=True)

replacetime = {'perm.' : 'permanent','contr.' : 'contract'}
dataset2['Contract Time'].replace(replacetime,inplace = True)
dataset2['Contract Time'].fillna('non_specified',inplace=True)
```

```python
dataset2['Contract Type'].value_counts()
```

```python
dataset2['Contract Time'].value_counts()
```

**Change category**

```python
dataset2.Category.replace({'Finance & Accounting Jobs': 'Accounting & Finance Jobs'},inplace=True)
dataset2.Category.value_counts()
```

**Change Salary**

Adding a new column to store the `Salary per annum`

```python
dataset2['Salary per annum'] = dataset2['Salary per month'].apply(lambda x: x * 12) #salary per annual 
```

**Change Date**

create another two columns to store date format.

```python
dataset2['OpenDate_parsed'] = pd.to_datetime(dataset2.OpenDate,format="%Y-%m-%d %H:%M:%S",errors='coerce')
dataset2['CloseDate_parsed'] = pd.to_datetime(dataset2.CloseDate,format="%Y-%m-%d %H:%M:%S",errors='coerce')
```

Replace date by using strftime.

```python
dataset2['OpenDate'] = dataset2['OpenDate_parsed'].apply(lambda x: x.strftime("%Y%m%dT%H%M%S"))
```

```python
dataset2['CloseDate'] = dataset2['CloseDate_parsed'].apply(lambda x: x.strftime("%Y%m%dT%H%M%S"))
```

```python
dataset2.head()
```

Drop the cloumns we crated before, and column we do not need.

```python
dataset2.drop(columns=['OpenDate_parsed', 'CloseDate_parsed','Salary per month'],inplace=True)
```

```python
dataset2.head()
```

Rename the columns with different names.

```python
dataset2 = dataset2.rename(columns = 
                 {'Source Name':'SourceName',
                  'Contract Type':'ContractType',
                  'Contract Time':'ContractTime',
                  'location':'Location'})
```

```python
dataset2.head()
```

In dataset1, we store the salary as string(contain range salary), we should also modify the type of salary in dataset2 as the string(or object).

```python
dataset2['Salary per annum'] = dataset2['Salary per annum'].astype(str) #fit data type with dataset1
```

```python
dataset2.info()
```

Try to merge two tables using outer join. on dataset1's column.

```python
mergedf = pd.merge(dataset1,dataset2,how='outer',on=dataset1.columns.tolist())
```

Check the dataset after we merge them togther.

```python
mergedf.describe(include='O')
```

From the above descibe function, we can found that there are some records have same `Title`, there may be possiblity that there are duplicate records in two datasets.
- - -
To identify which rows are duplicated, we should select a global key for the integrated data to decide what attribute make two records the duplicated record.
- - -
`Title` is one of the main attribute can identify record, however, there may be some different companies require similar job by using same title. We should use other attributes to restrict record. 
- - - 
Considering about the real world job advertisement, if we have a job in the same location, same company, with same category and salary, we can have high chance to identify unique job position.
- - -
In this case, we can add `Loacation`,`Category`,`Salary per annum`,`Company` as global key to identify duplicated records.

```python
mergedf[mergedf.duplicated(['Title','Location','Category','Salary per annum','Company'],keep = 'first')] #mark duplicates excpet first occur
```

Pick up one of the record as an example to verify our assumption

```python
mergedf[mergedf.Title == 'RGN/RMN']
```

Drop the diplicates records.

```python
mergedf.drop_duplicates(['Title','Location','Category','Salary per annum','Company'],inplace=True) # drop duplicate records
```

```python
mergedf.reset_index(inplace=True,drop=True) #reset the index.
```
